```python
# Import packages
import requests
from bs4 import BeautifulSoup
import textwrap
```

```python
# Get a URLs HTML soup
url = 'https://www.washingtonpost.com/science/2021/10/09/tiktok-jonsbones-human-remains-wall-spines/?itid=hp-top-table-high'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# job_elems = soup.find_all('p', class_='font--body font-copy gray-darkest ma-0 pb-md')  # didn't work for article on 8/8/21
job_elems = soup.find_all('p')

wrapper = textwrap.TextWrapper(width=100)
word_list = wrapper.wrap(text=' '.join([e.text for e in job_elems]))

for elem in word_list:
    obj = elem + '\n\n'
    if obj == '\n\n' or obj == '\n':
        print('')
    else:
        print(obj)
```

```python
for sent in soup.get_text(separator="\n").split('\n'):
    print(sent)
```
